## Scrapping for whole page

```python
name = []
rating = []
discounted_price = []
orignal_price = []
discount = []
description = []

# Considering whole container on the page
container = soup.find_all('div',class_='_2kHMtA')

for i in container:
    name.append(i.find('div',class_='_4rR01T').text.strip())
    rating.append(i.find('div',class_='_3LWZlK').text.strip())
    discounted_price.append(i.find('div',class_='_30jeq3 _1_WHN1').text.strip())
    orignal_price.append(i.find('div',class_='_3I9_wc _27UcVY').text.strip())
    discount.append(i.find('div',class_='_3Ay6Sb').text.strip())
    description.append(i.find('div',class_='fMghEO').text.strip())
    
df = pd.DataFrame({
    'name' : name,
    'rating' : rating,
    'discounted_price' : discounted_price,
    'orignal_price' : orignal_price,
    'discount' : discount,
    'description' : description

})
```

## Scrapping Every Page

```python
# for every page
final = pd.DataFrame()
for i in range(1,51):
    
    webpage1=requests.get(f'https://www.flipkart.com/search?q=mobiles&page={i}',headers = headers).text
    soup=BeautifulSoup(webpage1,'lxml')
    name = []
    rating = []
    discounted_price = []
    orignal_price = []
    discount = []
    description = []

    # Considering whole container on the page
    container = soup.find_all('div',class_='_2kHMtA')

    for j in container:
        try:
            name.append(j.find('div',class_='_4rR01T').text.strip())
        except:
            name.append(np.nan)
        try:    
            rating.append(j.find('div',class_='_3LWZlK').text.strip())
        except:
            rating.append(np.nan)
        try:
            discounted_price.append(j.find('div',class_='_30jeq3 _1_WHN1').text.strip())
        except:
            discounted_price.append(np.nan)
        try:
            orignal_price.append(j.find('div',class_='_3I9_wc _27UcVY').text.strip())
        except:
            orignal_price.append(np.nan)
        try:
            discount.append(j.find('div',class_='_3Ay6Sb').text.strip())
        except:
            discount.append(np.nan)
        try:
            description.append(j.find('div',class_='fMghEO').text.strip())
        except:
            description.append(np.nan)
        
        
    df = pd.DataFrame({
        'name' : name,
        'rating' : rating,
        'discounted_price' : discounted_price,
        'orignal_price' : orignal_price,
        'discount' : discount,
        'description' : description

        })
    final=final.append(df,ignore_index=True)
```
